```python
import numpy as np
import pandas as pd
```

```python
movie_names = pd.read_csv('movies.csv')
movie_names.head()
```

```python
rating_data = pd.read_csv('ratings.csv')
rating_data.head()
```

```python
movie_names['userId'] = pd.merge(movie_names, rating_data,on='movieId')['userId']
movie_names
```

```python
movie_data = pd.merge(rating_data, movie_names, on=['movieId', 'userId'])
movie_data.head()
```

### criteria for popularity based algorithm

1) most views
2) highest rating

```python
movie_data.groupby('title').mean().sort_values(by='rating', ascending=False).head(30)
```

```python
movie_data.groupby('title').count().sort_values(by='rating', ascending=False).head(10)
```

```python
ratings_mean_count = pd.DataFrame(movie_data.groupby(by='title')['rating'].mean())
```

```python
ratings_mean_count.head()
```

```python
ratings_mean_count['rating counts']=pd.DataFrame(movie_data.groupby('title')['rating'].count())

ratings_mean_count.head()
```

```python
ratings_mean_count['rating'] = round(ratings_mean_count['rating'],1)
```

```python
ratings_mean_count
```

```python
ratings_mean_count = ratings_mean_count[(ratings_mean_count['rating'] > 3 )& (ratings_mean_count['rating counts'] > 100)]
```

```python
ratings_mean_count.head()
```

```python
first_10_movies = ratings_mean_count.sort_values('rating', ascending=False).head(10)
```

```python
first_10_movies
```
